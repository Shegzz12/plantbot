```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Generates a complete plant care dataset for all 5-parameter combinations (5^5 = 3125 rows).
For each row it produces:
- Human-friendly bands for each parameter
- An overall description summarizing all five conditions
- A status label (Optimal / Good / Attention / Critical)
- Up to 10 refined, prioritized and non-duplicative actions covering ALL parameters
  (soil moisture, air humidity, air temperature, plant temperature, soil temperature)
"""

import csv
import itertools
import os
from typing import List, Tuple, Dict, Set

# -----------------------------
# User-specified ranges (exact)
# -----------------------------
RANGES: Dict[str, List[Tuple[int, int]]] = {
    "soil_moist": [(0, 20), (21, 40), (41, 60), (61, 80), (81, 100)],      # %
    "air_hum":    [(0, 20), (21, 40), (41, 60), (61, 80), (81, 100)],      # %
    "air_temp":   [(0, 10), (11, 20), (21, 30), (31, 40), (41, 100)],      # °C
    "plant_temp": [(15, 20), (21, 25), (26, 30), (31, 35), (36, 100)],     # °C
    "soil_temp":  [(15, 20), (21, 25), (26, 30), (31, 35), (36, 100)]      # °C
}

# -----------------------------
# Friendly labels per band
# -----------------------------
LABELS_SHORT = {
    "soil_moist": ["very dry", "dry", "moderate", "moist", "waterlogged"],
    "air_hum":    ["very low", "low", "moderate", "high", "very high"],
    "air_temp":   ["very cold", "cool", "comfortable", "hot", "very hot"],
    "plant_temp": ["slightly cool", "favourable", "optimal", "hot", "very hot"],
    "soil_temp":  ["cool", "optimal", "warm", "hot", "very hot"],
}

LABELS_LONG = {
    "soil_moist": [
        "Soil moisture is critically low",
        "Soil moisture is below ideal",
        "Soil moisture is within the optimal range",
        "Soil is quite wet",
        "Soil is saturated/waterlogged",
    ],
    "air_hum": [
        "Air humidity is critically low",
        "Air humidity is below ideal",
        "Air humidity is within the optimal range",
        "Air humidity is high",
        "Air humidity is excessively high",
    ],
    "air_temp": [
        "Ambient air temperature is too low",
        "Ambient air temperature is on the cool side",
        "Ambient air temperature is within the optimal range",
        "Ambient air temperature is high",
        "Ambient air temperature is dangerously high",
    ],
    "plant_temp": [
        "Leaf temperature is slightly cool",
        "Leaf temperature is favourable",
        "Leaf temperature is optimal for photosynthesis",
        "Leaf temperature is high (heat stress risk)",
        "Leaf temperature is very high (severe heat stress risk)",
    ],
    "soil_temp": [
        "Soil temperature is cool",
        "Soil temperature is optimal for root activity",
        "Soil temperature is warm",
        "Soil temperature is hot (root stress risk)",
        "Soil temperature is very hot (root damage risk)",
    ],
}

# -----------------------------
# Optimal band(s) per parameter
# (sets allow a param to have multiple optimal bands)
# -----------------------------
OPTIMAL_IDX: Dict[str, Set[int]] = {
    "soil_moist": {2},        # 41–60% (moderate)
    "air_hum":    {2},        # 41–60% (moderate)
    "air_temp":   {2},        # 21–30°C (comfortable)
    "plant_temp": {2},        # 26–30°C (optimal)
    "soil_temp":  {1, 2},     # 21–25°C or 26–30°C acceptable/optimal for many crops
}

# -----------------------------
# Severity scoring (0 best → 3 worst)
# -----------------------------
def band_severity(param: str, idx: int) -> int:
    """Assign a severity level to a band index per parameter."""
    # Generic mapping around the optimal band(s)
    if idx in OPTIMAL_IDX[param]:
        return 0
    # Mild deviations (one step away)
    if any(abs(idx - o) == 1 for o in OPTIMAL_IDX[param]):
        return 1
    # Moderate deviations (two steps away)
    if any(abs(idx - o) == 2 for o in OPTIMAL_IDX[param]):
        return 2
    # Severe deviations (>= three steps away)
    return 3

# -----------------------------
# Helpers
# -----------------------------
def band_str(bounds: Tuple[int, int], unit: str) -> str:
    lo, hi = bounds
    if unit == "%":
        return f"{lo}–{hi}%"
    if unit == "°C":
        return f"{lo}–{hi}°C" if hi < 100 else f"{lo}°C and above"
    return f"{lo}–{hi}"

def overall_status(severities: Dict[str, int]) -> str:
    s = max(severities.values())
    if s == 0:
        return "Optimal"
    if s == 1:
        return "Good"
    if s == 2:
        return "Attention"
    return "Critical"

# -----------------------------
# Per-parameter actions (by band index)
# These produce brief, actionable, *non-duplicative* steps.
# -----------------------------
def actions_soil_moist(idx: int) -> List[str]:
    return [
        "Irrigate deeply now; target root zone and avoid runoff.",
        "Increase irrigation slightly; add organic mulch to reduce evaporation.",
        "Maintain current irrigation schedule; spot-check with a moisture probe weekly.",
        "Reduce watering; improve aeration and check that drainage holes are clear.",
        "Pause irrigation; create drainage paths or raised beds and loosen compacted soil.",
    ][idx:idx+1]

def actions_air_hum(idx: int) -> List[str]:
    return [
        "Raise humidity with light misting or humidity trays; avoid strong, drying airflow.",
        "Increase humidity gradually and group plants to create a local microclimate.",
        "Maintain steady airflow; avoid prolonged leaf wetness to deter disease.",
        "Improve cross-ventilation and widen spacing; avoid evening leaf wetting.",
        "Use ventilation/dehumidification; remove standing water and prune for airflow.",
    ][idx:idx+1]

def actions_air_temp(idx: int) -> List[str]:
    return [
        "Protect from cold using row covers, windbreaks, or relocate to warmer spot.",
        "Use light covers at night and avoid cold irrigation; monitor for slow growth.",
        "Maintain current exposure; ventilate greenhouses during midday.",
        "Provide shade during peak sun, irrigate early morning, and avoid midday pruning.",
        "Deploy emergency shade cloth/evaporative cooling; prioritize plant survival.",
    ][idx:idx+1]

def actions_plant_temp(idx: int) -> List[str]:
    return [
        "Increase morning sun exposure and ensure soil is warm; reduce excessive shade.",
        "Maintain exposure; fine-tune irrigation to support steady growth.",
        "Conditions are ideal for photosynthesis; keep nutrition on schedule.",
        "Provide partial shade and intermittent misting; prevent leaf scorch.",
        "Aggressively cool canopy (dense shade, frequent light mist in cooler hours).",
    ][idx:idx+1]

def actions_soil_temp(idx: int) -> List[str]:
    return [
        "Warm the root zone with dark mulch or low tunnels; avoid overwatering cool soil.",
        "Maintain mulch cover; keep soil evenly moist for strong root activity.",
        "Buffer heat swings with mulch; irrigate mornings to avoid hot soil wetting.",
        "Shade the soil surface; increase mulch thickness and check root health.",
        "Cool the soil with heavy mulch and deep morning irrigation; consider reflective mulch.",
    ][idx:idx+1]

# -----------------------------
# Synergy actions (cross-parameter)
# -----------------------------
def synergy_actions(sm_i, ah_i, at_i, pt_i, st_i) -> List[str]:
    out: List[str] = []
    # Drought + hot air or hot plant → urgent cooling/rehydration
    if sm_i in (0, 1) and (at_i in (3, 4) or pt_i in (3, 4)):
        out.append("Prioritize deep irrigation and immediate shading; pause fertiliser until stress subsides.")
    # Waterlogged + high humidity → pathogen pressure
    if sm_i in (3, 4) and ah_i in (3, 4):
        out.append("High disease risk: improve drainage and airflow; remove dead leaves promptly.")
    # Cool soil + wet → root-rot risk
    if st_i == 0 and sm_i in (3, 4):
        out.append("Cool, wet soil slows respiration: drain excess water and gently warm soil with dark mulch.")
    # Very hot soil + dry → root scorch
    if st_i == 4 and sm_i in (0, 1):
        out.append("Roots at risk of scorch: deep water then mulch heavily; consider reflective ground cover.")
    # Very low humidity + hot leaves → excessive transpiration/VPD
    if ah_i == 0 and pt_i in (3, 4):
        out.append("Reduce VPD: provide intermittent mist and partial shade to stabilise leaf water loss.")
    # Very hot air + very hot soil → system-wide heat stress
    if at_i == 4 and st_i in (3, 4):
        out.append("System-wide heat: combine shade cloth, ventilation, and early-morning irrigation.")
    return out

# -----------------------------
# Primary headline (soil moisture + plant temperature)
# -----------------------------
def headline_soil_plant(sm_i: int, pt_i: int) -> str:
    # Map to concise combined headline
    soil_map = {0: "Soil very dry", 1: "Soil dry", 2: "Soil optimal moisture", 3: "Soil wet", 4: "Soil waterlogged"}
    plant_map = {0: "plant slightly cool", 1: "plant favourable temp", 2: "plant optimal temp", 3: "plant hot", 4: "plant very hot"}
    return f"{soil_map[sm_i]}, {plant_map[pt_i]}"

# -----------------------------
# Overall description builder
# -----------------------------
def build_overall_description(indices: Dict[str, int]) -> str:
    parts = []
    # In a single, flowing sentence:
    parts.append(
        f"Soil moisture is {LABELS_SHORT['soil_moist'][indices['soil_moist']]} "
        f"({band_str(RANGES['soil_moist'][indices['soil_moist']], '%')}), "
        f"air humidity is {LABELS_SHORT['air_hum'][indices['air_hum']]} "
        f"({band_str(RANGES['air_hum'][indices['air_hum']], '%')}), "
        f"air temperature is {LABELS_SHORT['air_temp'][indices['air_temp']]} "
        f"({band_str(RANGES['air_temp'][indices['air_temp']], '°C')}), "
        f"leaf temperature is {LABELS_SHORT['plant_temp'][indices['plant_temp']]} "
        f"({band_str(RANGES['plant_temp'][indices['plant_temp']], '°C')}), "
        f"and soil temperature is {LABELS_SHORT['soil_temp'][indices['soil_temp']]} "
        f"({band_str(RANGES['soil_temp'][indices['soil_temp']], '°C')})."
    )
    return "".join(parts)

# -----------------------------
# Action aggregator & prioritiser
# -----------------------------
def gather_actions(sm_i, ah_i, at_i, pt_i, st_i, status: str) -> List[str]:
    actions: List[str] = []

    # 1) Synergy first (cross-parameter)
    actions.extend(synergy_actions(sm_i, ah_i, at_i, pt_i, st_i))

    # 2) Primary headline context-specific
    headline = headline_soil_plant(sm_i, pt_i)
    if "optimal" not in headline.lower():  # add only if there's a non-optimal note
        actions.insert(0, headline + " → address this first.")

    # 3) Parameter-specific actions (ordered by severity)
    severities = {
        "soil_moist": band_severity("soil_moist", sm_i),
        "air_hum":    band_severity("air_hum", ah_i),
        "air_temp":   band_severity("air_temp", at_i),
        "plant_temp": band_severity("plant_temp", pt_i),
        "soil_temp":  band_severity("soil_temp", st_i),
    }
    sorted_params = sorted(severities.items(), key=lambda kv: kv[1], reverse=True)

    for param, _sev in sorted_params:
        idx = {"soil_moist": sm_i, "air_hum": ah_i, "air_temp": at_i, "plant_temp": pt_i, "soil_temp": st_i}[param]
        if param == "soil_moist":
            actions.extend(actions_soil_moist(idx))
        elif param == "air_hum":
            actions.extend(actions_air_hum(idx))
        elif param == "air_temp":
            actions.extend(actions_air_temp(idx))
        elif param == "plant_temp":
            actions.extend(actions_plant_temp(idx))
        elif param == "soil_temp":
            actions.extend(actions_soil_temp(idx))

    # 4) Optimisation/maintenance layer
    if status == "Optimal":
        actions = [
            "All parameters are optimal: maintain irrigation and nutrition schedules.",
            "Scout weekly for pests/disease; act early if symptoms appear.",
            "Keep mulch layer 3–5 cm to stabilise soil moisture and temperature.",
            "Ventilate during midday to prevent sudden humidity spikes.",
            "Record conditions and yields to fine-tune future irrigation/fertility.",
        ] + actions  # keep param-specific context after maintenance tips

    # Deduplicate while preserving order
    seen = set()
    unique_actions = []
    for a in actions:
        if a and a not in seen:
            seen.add(a)
            unique_actions.append(a)

    # Cap at 10
    return unique_actions[:10]

# -----------------------------
# Main generator
# -----------------------------
def generate_csv(out_path: str) -> None:
    headers = [
        "SoilMoistureBand", "AirHumidityBand", "AirTempBand", "PlantTempBand", "SoilTempBand",
        "Status", "OverallDescription",
        "Advice1", "Advice2", "Advice3", "Advice4", "Advice5",
        "Advice6", "Advice7", "Advice8", "Advice9", "Advice10"
    ]

    rows: List[List[str]] = []

    for sm_i, ah_i, at_i, pt_i, st_i in itertools.product(range(5), repeat=5):
        indices = {"soil_moist": sm_i, "air_hum": ah_i, "air_temp": at_i, "plant_temp": pt_i, "soil_temp": st_i}

        # Bands as readable strings
        sm_band = band_str(RANGES["soil_moist"][sm_i], "%")
        ah_band = band_str(RANGES["air_hum"][ah_i], "%")
        at_band = band_str(RANGES["air_temp"][at_i], "°C")
        pt_band = band_str(RANGES["plant_temp"][pt_i], "°C")
        st_band = band_str(RANGES["soil_temp"][st_i], "°C")

        # Status
        severities = {p: band_severity(p, idx) for p, idx in indices.items()}
        status = overall_status(severities)

        # Overall description
        overall_desc = build_overall_description(indices)

        # Actions (always produced; cover all 5 parameters)
        advices = gather_actions(sm_i, ah_i, at_i, pt_i, st_i, status)
        while len(advices) < 10:
            advices.append("")

        row = [
            sm_band, ah_band, at_band, pt_band, st_band,
            status, overall_desc,
            *advices[:10],
        ]
        rows.append(row)

    # Write CSV
    with open(out_path, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(headers)
        writer.writerows(rows)

    print(f"✔ Generated {len(rows)} rows (should be 3125).")
    print(f"Saved to: {os.path.abspath(out_path)}")

# -----------------------------
# Entry point
# -----------------------------
if __name__ == "__main__":
    output_file = "plant_health_advice_full_3125.csv"
    generate_csv(output_file)
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Generates a complete plant care dataset for all 5-parameter combinations (5^5 = 3125 rows).
For each row it produces:
- Human-friendly bands for each parameter (kept for descriptions)
- Numeric category fields (1..5) for each input (soil_moist, air_hum, air_temp, plant_temp, soil_temp)
- An overall description summarizing all five conditions
- A status label (Optimal / Good / Attention / Critical)
- Up to 10 refined, prioritized and non-duplicative actions covering ALL parameters
  (soil moisture, air humidity, air temperature, plant temperature, soil temperature)
"""

import csv
import itertools
import os
from typing import List, Tuple, Dict, Set

# -----------------------------
# User-specified ranges (exact)
# -----------------------------
RANGES: Dict[str, List[Tuple[int, int]]] = {
    "soil_moist": [(0, 20), (21, 40), (41, 60), (61, 80), (81, 100)],      # %
    "air_hum":    [(0, 20), (21, 40), (41, 60), (61, 80), (81, 100)],      # %
    "air_temp":   [(0, 10), (11, 20), (21, 30), (31, 40), (41, 100)],      # °C
    "plant_temp": [(15, 20), (21, 25), (26, 30), (31, 35), (36, 100)],     # °C
    "soil_temp":  [(15, 20), (21, 25), (26, 30), (31, 35), (36, 100)]      # °C
}

# -----------------------------
# Friendly labels per band
# -----------------------------
LABELS_SHORT = {
    "soil_moist": ["very dry", "dry", "moderate", "moist", "waterlogged"],
    "air_hum":    ["very low", "low", "moderate", "high", "very high"],
    "air_temp":   ["very cold", "cool", "comfortable", "hot", "very hot"],
    "plant_temp": ["slightly cool", "favourable", "optimal", "hot", "very hot"],
    "soil_temp":  ["cool", "optimal", "warm", "hot", "very hot"],
}

LABELS_LONG = {
    "soil_moist": [
        "Soil moisture is critically low",
        "Soil moisture is below ideal",
        "Soil moisture is within the optimal range",
        "Soil is quite wet",
        "Soil is saturated/waterlogged",
    ],
    "air_hum": [
        "Air humidity is critically low",
        "Air humidity is below ideal",
        "Air humidity is within the optimal range",
        "Air humidity is high",
        "Air humidity is excessively high",
    ],
    "air_temp": [
        "Ambient air temperature is too low",
        "Ambient air temperature is on the cool side",
        "Ambient air temperature is within the optimal range",
        "Ambient air temperature is high",
        "Ambient air temperature is dangerously high",
    ],
    "plant_temp": [
        "Leaf temperature is slightly cool",
        "Leaf temperature is favourable",
        "Leaf temperature is optimal for photosynthesis",
        "Leaf temperature is high (heat stress risk)",
        "Leaf temperature is very high (severe heat stress risk)",
    ],
    "soil_temp": [
        "Soil temperature is cool",
        "Soil temperature is optimal for root activity",
        "Soil temperature is warm",
        "Soil temperature is hot (root stress risk)",
        "Soil temperature is very hot (root damage risk)",
    ],
}

# -----------------------------
# Optimal band(s) per parameter
# (sets allow a param to have multiple optimal bands)
# -----------------------------
OPTIMAL_IDX: Dict[str, Set[int]] = {
    "soil_moist": {2},        # 41–60% (moderate)
    "air_hum":    {2},        # 41–60% (moderate)
    "air_temp":   {2},        # 21–30°C (comfortable)
    "plant_temp": {2},        # 26–30°C (optimal)
    "soil_temp":  {1, 2},     # 21–25°C or 26–30°C acceptable/optimal for many crops
}

# -----------------------------
# Severity scoring (0 best → 3 worst)
# -----------------------------
def band_severity(param: str, idx: int) -> int:
    """Assign a severity level to a band index per parameter (idx is 0-based band index)."""
    if idx in OPTIMAL_IDX[param]:
        return 0
    if any(abs(idx - o) == 1 for o in OPTIMAL_IDX[param]):
        return 1
    if any(abs(idx - o) == 2 for o in OPTIMAL_IDX[param]):
        return 2
    return 3

# -----------------------------
# Helpers
# -----------------------------
def band_str(bounds: Tuple[int, int], unit: str) -> str:
    """Return a readable band string using ASCII hyphen to avoid encoding issues."""
    lo, hi = bounds
    if unit == "%":
        return f"{lo}-{hi}%"
    if unit == "°C":
        return f"{lo}-{hi}°C" if hi < 100 else f"{lo}°C and above"
    return f"{lo}-{hi}"

def overall_status(severities: Dict[str, int]) -> str:
    s = max(severities.values())
    if s == 0:
        return "Optimal"
    if s == 1:
        return "Good"
    if s == 2:
        return "Attention"
    return "Critical"

# -----------------------------
# Map raw numeric sensor value -> category number 1..5
# -----------------------------
def value_to_category(param: str, value: float) -> int:
    """
    Convert a raw sensor value to a category number (1..5) using RANGES.
    Returns 1..5 (1 = lowest band, 5 = highest band).
    If value is lower than the first band's low, returns 1.
    If value >= the start of the last band, returns 5.
    """
    lst = RANGES[param]
    # get starts for each band
    starts = [low for (low, _hi) in lst]
    # iterate from highest to lowest start: first start <= value -> that band
    for i in range(len(starts) - 1, -1, -1):
        if value >= starts[i]:
            return i + 1
    return 1

# -----------------------------
# Per-parameter actions (by band index)
# These produce brief, actionable, *non-duplicative* steps.
# (unchanged)
# -----------------------------
def actions_soil_moist(idx: int) -> List[str]:
    return [
        "Irrigate deeply now; target root zone and avoid runoff.",
        "Increase irrigation slightly; add organic mulch to reduce evaporation.",
        "Maintain current irrigation schedule; spot-check with a moisture probe weekly.",
        "Reduce watering; improve aeration and check that drainage holes are clear.",
        "Pause irrigation; create drainage paths or raised beds and loosen compacted soil.",
    ][idx:idx+1]

def actions_air_hum(idx: int) -> List[str]:
    return [
        "Raise humidity with light misting or humidity trays; avoid strong, drying airflow.",
        "Increase humidity gradually and group plants to create a local microclimate.",
        "Maintain steady airflow; avoid prolonged leaf wetness to deter disease.",
        "Improve cross-ventilation and widen spacing; avoid evening leaf wetting.",
        "Use ventilation/dehumidification; remove standing water and prune for airflow.",
    ][idx:idx+1]

def actions_air_temp(idx: int) -> List[str]:
    return [
        "Protect from cold using row covers, windbreaks, or relocate to warmer spot.",
        "Use light covers at night and avoid cold irrigation; monitor for slow growth.",
        "Maintain current exposure; ventilate greenhouses during midday.",
        "Provide shade during peak sun, irrigate early morning, and avoid midday pruning.",
        "Deploy emergency shade cloth/evaporative cooling; prioritize plant survival.",
    ][idx:idx+1]

def actions_plant_temp(idx: int) -> List[str]:
    return [
        "Increase morning sun exposure and ensure soil is warm; reduce excessive shade.",
        "Maintain exposure; fine-tune irrigation to support steady growth.",
        "Conditions are ideal for photosynthesis; keep nutrition on schedule.",
        "Provide partial shade and intermittent misting; prevent leaf scorch.",
        "Aggressively cool canopy (dense shade, frequent light mist in cooler hours).",
    ][idx:idx+1]

def actions_soil_temp(idx: int) -> List[str]:
    return [
        "Warm the root zone with dark mulch or low tunnels; avoid overwatering cool soil.",
        "Maintain mulch cover; keep soil evenly moist for strong root activity.",
        "Buffer heat swings with mulch; irrigate mornings to avoid hot soil wetting.",
        "Shade the soil surface; increase mulch thickness and check root health.",
        "Cool the soil with heavy mulch and deep morning irrigation; consider reflective mulch.",
    ][idx:idx+1]

# -----------------------------
# Synergy actions (cross-parameter)
# -----------------------------
def synergy_actions(sm_i, ah_i, at_i, pt_i, st_i) -> List[str]:
    out: List[str] = []
    # Drought + hot air or hot plant → urgent cooling/rehydration
    if sm_i in (0, 1) and (at_i in (3, 4) or pt_i in (3, 4)):
        out.append("Prioritize deep irrigation and immediate shading; pause fertiliser until stress subsides.")
    # Waterlogged + high humidity → pathogen pressure
    if sm_i in (3, 4) and ah_i in (3, 4):
        out.append("High disease risk: improve drainage and airflow; remove dead leaves promptly.")
    # Cool soil + wet → root-rot risk
    if st_i == 0 and sm_i in (3, 4):
        out.append("Cool, wet soil slows respiration: drain excess water and gently warm soil with dark mulch.")
    # Very hot soil + dry → root scorch
    if st_i == 4 and sm_i in (0, 1):
        out.append("Roots at risk of scorch: deep water then mulch heavily; consider reflective ground cover.")
    # Very low humidity + hot leaves → excessive transpiration/VPD
    if ah_i == 0 and pt_i in (3, 4):
        out.append("Reduce VPD: provide intermittent mist and partial shade to stabilise leaf water loss.")
    # Very hot air + very hot soil → system-wide heat stress
    if at_i == 4 and st_i in (3, 4):
        out.append("System-wide heat: combine shade cloth, ventilation, and early-morning irrigation.")
    return out

# -----------------------------
# Primary headline (soil moisture + plant temperature)
# -----------------------------
def headline_soil_plant(sm_i: int, pt_i: int) -> str:
    soil_map = {0: "Soil very dry", 1: "Soil dry", 2: "Soil optimal moisture", 3: "Soil wet", 4: "Soil waterlogged"}
    plant_map = {0: "plant slightly cool", 1: "plant favourable temp", 2: "plant optimal temp", 3: "plant hot", 4: "plant very hot"}
    return f"{soil_map[sm_i]}, {plant_map[pt_i]}"

# -----------------------------
# Overall description builder
# -----------------------------
def build_overall_description(indices: Dict[str, int]) -> str:
    parts = []
    parts.append(
        f"Soil moisture is {LABELS_SHORT['soil_moist'][indices['soil_moist']]} "
        f"({band_str(RANGES['soil_moist'][indices['soil_moist']], '%')}), "
        f"air humidity is {LABELS_SHORT['air_hum'][indices['air_hum']]} "
        f"({band_str(RANGES['air_hum'][indices['air_hum']], '%')}), "
        f"air temperature is {LABELS_SHORT['air_temp'][indices['air_temp']]} "
        f"({band_str(RANGES['air_temp'][indices['air_temp']], '°C')}), "
        f"leaf temperature is {LABELS_SHORT['plant_temp'][indices['plant_temp']]} "
        f"({band_str(RANGES['plant_temp'][indices['plant_temp']], '°C')}), "
        f"and soil temperature is {LABELS_SHORT['soil_temp'][indices['soil_temp']]} "
        f"({band_str(RANGES['soil_temp'][indices['soil_temp']], '°C')})."
    )
    return "".join(parts)

# -----------------------------
# Action aggregator & prioritiser
# -----------------------------
def gather_actions(sm_i, ah_i, at_i, pt_i, st_i, status: str) -> List[str]:
    actions: List[str] = []

    # 1) Synergy first (cross-parameter)
    actions.extend(synergy_actions(sm_i, ah_i, at_i, pt_i, st_i))

    # 2) Primary headline context-specific
    headline = headline_soil_plant(sm_i, pt_i)
    if "optimal" not in headline.lower():  # add only if there's a non-optimal note
        actions.insert(0, headline + " → address this first.")

    # 3) Parameter-specific actions (ordered by severity)
    severities = {
        "soil_moist": band_severity("soil_moist", sm_i),
        "air_hum":    band_severity("air_hum", ah_i),
        "air_temp":   band_severity("air_temp", at_i),
        "plant_temp": band_severity("plant_temp", pt_i),
        "soil_temp":  band_severity("soil_temp", st_i),
    }
    sorted_params = sorted(severities.items(), key=lambda kv: kv[1], reverse=True)

    for param, _sev in sorted_params:
        idx = {"soil_moist": sm_i, "air_hum": ah_i, "air_temp": at_i, "plant_temp": pt_i, "soil_temp": st_i}[param]
        if param == "soil_moist":
            actions.extend(actions_soil_moist(idx))
        elif param == "air_hum":
            actions.extend(actions_air_hum(idx))
        elif param == "air_temp":
            actions.extend(actions_air_temp(idx))
        elif param == "plant_temp":
            actions.extend(actions_plant_temp(idx))
        elif param == "soil_temp":
            actions.extend(actions_soil_temp(idx))

    # 4) Optimisation/maintenance layer
    if status == "Optimal":
        actions = [
            "All parameters are optimal: maintain irrigation and nutrition schedules.",
            "Scout weekly for pests/disease; act early if symptoms appear.",
            "Keep mulch layer 3–5 cm to stabilise soil moisture and temperature.",
            "Ventilate during midday to prevent sudden humidity spikes.",
            "Record conditions and yields to fine-tune future irrigation/fertility.",
        ] + actions  # keep param-specific context after maintenance tips

    # Deduplicate while preserving order
    seen = set()
    unique_actions = []
    for a in actions:
        if a and a not in seen:
            seen.add(a)
            unique_actions.append(a)

    # Cap at 10
    return unique_actions[:10]

# -----------------------------
# Main generator
# -----------------------------
def generate_csv(out_path: str) -> None:
    headers = [
        # First five fields are numeric categories 1..5 (not textual ranges)
        "SoilMoistureCat", "AirHumidityCat", "AirTempCat", "PlantTempCat", "SoilTempCat",
        "Status", "OverallDescription",
        "Advice1", "Advice2", "Advice3", "Advice4", "Advice5",
        "Advice6", "Advice7", "Advice8", "Advice9", "Advice10"
    ]

    rows: List[List[str]] = []

    for sm_i, ah_i, at_i, pt_i, st_i in itertools.product(range(5), repeat=5):
        indices = {"soil_moist": sm_i, "air_hum": ah_i, "air_temp": at_i, "plant_temp": pt_i, "soil_temp": st_i}

        # -----------------------------
        # Numeric category fields (1..5) as requested
        # -----------------------------
        sm_cat = sm_i + 1
        ah_cat = ah_i + 1
        at_cat = at_i + 1
        pt_cat = pt_i + 1
        st_cat = st_i + 1

        # Status
        severities = {p: band_severity(p, idx) for p, idx in indices.items()}
        status = overall_status(severities)

        # Overall description (still uses human-friendly band strings internally)
        overall_desc = build_overall_description(indices)

        # Actions (always produced; cover all 5 parameters)
        advices = gather_actions(sm_i, ah_i, at_i, pt_i, st_i, status)
        while len(advices) < 10:
            advices.append("")

        row = [
            sm_cat, ah_cat, at_cat, pt_cat, st_cat,
            status, overall_desc,
            *advices[:10],
        ]
        rows.append(row)

    # Write CSV (UTF-8, ASCII hyphens in descriptions)
    with open(out_path, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(headers)
        writer.writerows(rows)

    print(f"✔ Generated {len(rows)} rows (should be 3125).")
    print(f"Saved to: {os.path.abspath(out_path)}")

# -----------------------------
# Entry point
# -----------------------------
if __name__ == "__main__":
    output_file = "plant_health_advice_full_3126.csv"
    generate_csv(output_file)
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Generates a complete plant care dataset for all 5-parameter combinations (5^5 = 3125 rows).
For each row it produces:
- Numeric category fields (1..5) for each input (soil_moist, air_hum, air_temp, plant_temp, soil_temp)
- A single summarized description code (0..9) corresponding to one of 10 behavioral descriptions
- A single corresponding advice code (0..9) (kept aligned with the behavior code)
"""

import csv
import itertools
import os
from typing import List, Tuple, Dict, Set

# -----------------------------
# User-specified ranges (exact)
# -----------------------------
RANGES: Dict[str, List[Tuple[int, int]]] = {
    "soil_moist": [(0, 20), (21, 40), (41, 60), (61, 80), (81, 100)],      # %
    "air_hum":    [(0, 20), (21, 40), (41, 60), (61, 80), (81, 100)],      # %
    "air_temp":   [(0, 10), (11, 20), (21, 30), (31, 40), (41, 100)],      # °C
    "plant_temp": [(15, 20), (21, 25), (26, 30), (31, 35), (36, 100)],     # °C
    "soil_temp":  [(15, 20), (21, 25), (26, 30), (31, 35), (36, 100)]      # °C
}

# -----------------------------
# Friendly labels per band (kept for readable overall description)
# -----------------------------
LABELS_SHORT = {
    "soil_moist": ["very dry", "dry", "moderate", "moist", "waterlogged"],
    "air_hum":    ["very low", "low", "moderate", "high", "very high"],
    "air_temp":   ["very cold", "cool", "comfortable", "hot", "very hot"],
    "plant_temp": ["slightly cool", "favourable", "optimal", "hot", "very hot"],
    "soil_temp":  ["cool", "optimal", "warm", "hot", "very hot"],
}

# -----------------------------
# Optimal band(s) per parameter (0-based indices)
# -----------------------------
OPTIMAL_IDX: Dict[str, Set[int]] = {
    "soil_moist": {2},        # 41–60% (moderate)
    "air_hum":    {2},        # 41–60% (moderate)
    "air_temp":   {2},        # 21–30°C (comfortable)
    "plant_temp": {2},        # 26–30°C (optimal)
    "soil_temp":  {1, 2},     # 21–25°C or 26–30°C acceptable/optimal
}

# -----------------------------
# Severity scoring helper (kept)
# -----------------------------
def band_severity(param: str, idx: int) -> int:
    """Assign a severity level to a band index per parameter (0-based index)."""
    if idx in OPTIMAL_IDX[param]:
        return 0
    if any(abs(idx - o) == 1 for o in OPTIMAL_IDX[param]):
        return 1
    if any(abs(idx - o) == 2 for o in OPTIMAL_IDX[param]):
        return 2
    return 3

# -----------------------------
# Helpers
# -----------------------------
def band_str(bounds: Tuple[int, int], unit: str) -> str:
    """Return a readable band string using ASCII hyphen to avoid encoding issues."""
    lo, hi = bounds
    if unit == "%":
        return f"{lo}-{hi}%"
    if unit == "°C":
        return f"{lo}-{hi}°C" if hi < 100 else f"{lo}°C and above"
    return f"{lo}-{hi}"

def overall_status(severities: Dict[str, int]) -> str:
    s = max(severities.values())
    if s == 0:
        return "Optimal"
    if s == 1:
        return "Good"
    if s == 2:
        return "Attention"
    return "Critical"

# -----------------------------
# Build readable overall description (single sentence covering all five)
# (kept for internal use; we won't write this text to CSV)
# -----------------------------
def build_overall_description(indices: Dict[str, int]) -> str:
    return (
        f"Soil moisture: {LABELS_SHORT['soil_moist'][indices['soil_moist']]} ({band_str(RANGES['soil_moist'][indices['soil_moist']], '%')}), "
        f"Air humidity: {LABELS_SHORT['air_hum'][indices['air_hum']]} ({band_str(RANGES['air_hum'][indices['air_hum']], '%')}), "
        f"Air temp: {LABELS_SHORT['air_temp'][indices['air_temp']]} ({band_str(RANGES['air_temp'][indices['air_temp']], '°C')}), "
        f"Leaf temp: {LABELS_SHORT['plant_temp'][indices['plant_temp']]} ({band_str(RANGES['plant_temp'][indices['plant_temp']], '°C')}), "
        f"Soil temp: {LABELS_SHORT['soil_temp'][indices['soil_temp']]} ({band_str(RANGES['soil_temp'][indices['soil_temp']], '°C')})."
    )

# -----------------------------
# Define the 10 behavioral descriptions and matching advice (indices 0..9)
# (kept identical to your list)
# -----------------------------
BEHAVIOR_DESCRIPTIONS = [
    "All parameters optimal",
    "Dry soil with cool plant",
    "Dry soil with heat stress",
    "Waterlogged soil with cool plant",
    "Waterlogged soil with heat stress",
    "Low humidity combined with heat (transpiration stress)",
    "High humidity and waterlogging risk (disease-prone)",
    "Systemic heat stress (air + soil + leaves hot)",
    "Systemic cold stress (air/soil/leaf too cool)",
    "Minor/mixed deviations — small adjustments recommended"
]

BEHAVIOR_ADVICE = [
    "Conditions are within ideal ranges. Maintain irrigation and nutrition schedules, scout weekly for pests, keep mulch 3–5cm, and record conditions for tuning.",
    "Soil moisture is too low while leaf temps are cool: irrigate deeply to wet the root zone, add warm mulch, protect plants from cold at night using covers.",
    "Drought + heat: apply deep irrigation immediately, deploy shade cloth during peak sun, mist in cooler hours, and avoid midday foliar wetting to prevent burns.",
    "Waterlogged & cool: stop irrigation, improve surface drainage, gently aerate soil, apply dark mulch to raise soil temperature slightly and reduce root-rot risk.",
    "Waterlogged & hot: improve drainage immediately, reduce standing water, provide partial shade, pause further watering until soil drains, and monitor for disease.",
    "Low humidity with high temperatures: raise local humidity with intermittent misting or trays, add partial shade, and monitor plant turgor and VPD to avoid excessive transpiration.",
    "High humidity combined with wet soil: high disease risk — increase ventilation, space plants, remove standing water, and prune dense growth to improve airflow.",
    "System-wide heat stress: combine shade cloth, enhanced ventilation, deep early-morning irrigation, and consider evaporative cooling if available to protect canopy and roots.",
    "Systemic cold stress: protect plants at night with covers, minimize cold irrigation, consider temporary heating in protected structures, and delay transplanting until warmer.",
    "Minor or mixed deviations: make small targeted adjustments (slightly tweak irrigation, add thin mulch, introduce partial shade or a fan) and monitor daily for trends."
]

# -----------------------------
# Assign one of the 10 behaviors to each 5-field combination
# (kept the same logic)
# -----------------------------
def assign_behavior(sm_i: int, ah_i: int, at_i: int, pt_i: int, st_i: int) -> int:
    is_dry = sm_i in (0, 1)
    is_wet = sm_i in (3, 4)
    plant_cool = pt_i == 0
    plant_hot = pt_i in (3, 4)
    air_cold = at_i == 0
    air_hot = at_i in (3, 4)
    hum_low = ah_i == 0
    hum_high = ah_i == 4
    soil_hot = st_i in (3, 4)
    soil_cool = st_i == 0

    # 0: all optimal
    if (sm_i in OPTIMAL_IDX["soil_moist"] and
        ah_i in OPTIMAL_IDX["air_hum"] and
        at_i in OPTIMAL_IDX["air_temp"] and
        pt_i in OPTIMAL_IDX["plant_temp"] and
        st_i in OPTIMAL_IDX["soil_temp"]):
        return 0

    # 7: systemic heat stress
    if (air_hot and (soil_hot or plant_hot)):
        return 7

    # 8: systemic cold stress
    if (air_cold and (soil_cool or plant_cool)):
        return 8

    # 2: dry & heat
    if is_dry and plant_hot:
        return 2

    # 1: dry & cool
    if is_dry and plant_cool:
        return 1

    # 4: wet & hot
    if is_wet and plant_hot:
        return 4

    # 3: wet & cool
    if is_wet and plant_cool:
        return 3

    # 5: low humidity + heat
    if hum_low and (plant_hot or air_hot):
        return 5

    # 6: high humidity + wet
    if hum_high and is_wet:
        return 6

    # fallback: minor/mixed
    return 9

# -----------------------------
# Main generator: write numeric categories + behavior code + advice code
# -----------------------------
def generate_csv(out_path: str) -> None:
    headers = [
        "SoilMoistureCat", "AirHumidityCat", "AirTempCat", "PlantTempCat", "SoilTempCat",
        "SummaryDescription", "Advice"
    ]

    rows: List[List[str]] = []

    for sm_i, ah_i, at_i, pt_i, st_i in itertools.product(range(5), repeat=5):
        indices = {
            "soil_moist": sm_i,
            "air_hum": ah_i,
            "air_temp": at_i,
            "plant_temp": pt_i,
            "soil_temp": st_i
        }

        # Numeric categories 1..5
        sm_cat = sm_i + 1
        ah_cat = ah_i + 1
        at_cat = at_i + 1
        pt_cat = pt_i + 1
        st_cat = st_i + 1

        # Keep existing severity/status (not written)
        severities = {p: band_severity(p, idx) for p, idx in indices.items()}
        _status = overall_status(severities)

        # Build readable overall description internally (not written)
        _overall_desc = build_overall_description(indices)

        # Assign behavior index (0..9)
        beh_idx = assign_behavior(sm_i, ah_i, at_i, pt_i, st_i)

        # For numeric output: write the behavior index for both SummaryDescription and Advice
        rows.append([
            sm_cat, ah_cat, at_cat, pt_cat, st_cat,
            beh_idx,  # SummaryDescription numeric code (0..9)
            beh_idx   # Advice numeric code (0..9) — aligned with behavior
        ])

    # Write CSV with UTF-8 BOM so Excel reads symbols correctly (if any)
    with open(out_path, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow(headers)
        writer.writerows(rows)

    print(f"✔ Generated {len(rows)} rows (should be 3125).")
    print(f"Saved to: {os.path.abspath(out_path)}")

# -----------------------------
# Entry point
# -----------------------------
if __name__ == "__main__":
    output_file = "plant_health_advice_10patterns_codes_3125.csv"
    generate_csv(output_file)
```
